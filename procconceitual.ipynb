```python
from sys import exit
start = 'W'
endereco = {}
for place in range(0, 256): #Número de palavras endereçaveis = 2**8 = 256
    (endereco['{0:08b}'.format(place)]) = '' #A instrução deve estar 'vazia', uma vez que teria o C.Op. para parar o programa se fosse atríbuido '000000000000'
while start == 'W' or start == 'L' or start == 'R':
    start = input('Digite W para escrever, R para ler, L para listar toda a memoria e qualquer tecla para parar. ')
    if start == 'W':
        local = input('Digite o endereco de 8 bits: ') #Tamanho do endereço = 8 bits.
        palavra = input('Digite o dado de 12 bits: ') #Tamanho da célula e da palavvra = 12 bits
        endereco[local] = palavra
    elif start == 'R':
        local = input('Digite o endereco: ')
        print(endereco.get(local))
    elif start == 'L':
        print('Todos os estados: ')
        for element in endereco:
            print(endereco[element])
#Precisava de alguma forma escrever as instruções na memória :/
program_counter = '' 
acc_r0 = ''
for element2 in endereco:
    if endereco[element2] == '':
        continue
    else:
        program_counter = element2
        break
bits_c_op = 4   #c_op_tamanho = 4 bits
bits_c_operando = 8   #c_operando_tamanho = 8 bits    
while True:
    reg_end = program_counter #REM = CI
    program_counter = bin(int(program_counter, 2)+1)[2:].zfill(8)
    reg_dad = endereco.get(reg_end) #MP -> RDM
    reg_ins = reg_dad #RDM -> RI 
    c_op = ''.join(list(reg_ins)[0:bits_c_op])
    c_operando = ''.join(reg_ins[bits_c_op:]) #Op.(endereço)
#(x) = dado contido no endereço x; x = endereço x
    if c_op == '0000':
        exit(0) #Para a execução do programa
    elif c_op == '0001':
        acc_r0 = endereco[c_operando] #R0 <- (Op.)
    elif c_op == '0010':
        endereco[c_operando] = acc_r0 #(Op.) <- R0
    elif c_op == '0011':
        acc_r0 = bin((int(acc_r0, 2) + int(endereco[c_operando], 2)))[2:].zfill(12) #R0 = R0 + (Op.)
    elif c_op == '0100':
        acc_r0 = bin((int(acc_r0, 2) - int(endereco[c_operando], 2)))[2:].zfill(12) #R0 = R0 - (Op.)
    elif c_op == '0101':
        if int(acc_r0, 2) == 0: #Se R0 = 0:
            program_counter = c_operando #CI <- Op.
    elif c_op == '0110':
        if int(acc_r0, 2) > 0: #Se R0 > 0:
            program_counter = c_operando #CI <- Op.
    elif c_op == '0111':
        if int(acc_r0, 2) < 0: #Se R0 < 0:
            program_counter = c_operando #CI <- Op.
    elif c_op == '1000':
        program_counter = c_operando #CI <- Op.
    elif c_op == '1001':
        endereco[c_operando] = input('Digite o dado de 12 bits: ')
    elif c_op == '1010':
        print(endereco[c_operando])
```
